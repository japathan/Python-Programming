```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
pd.read_excel('supermarket.xlsx')
```

```python
df=pd.read_excel('supermarket.xlsx')
```

```python
df.head()
```

```python
df['CashierName'].unique()
```

```python
###  Mean ###
df1_mean=df.groupby(['CashierName'])['Amount','Received Cash'].mean().reset_index()
```

```python
df1_mean
```

```python
df1_sum=df.groupby(['CashierName'])['Amount'].sum().reset_index()
```

```python
df1_sum
```

```python
plt.figure(figsize=(25,10))
sns.lineplot(x='CashierName',y='Amount',data=df1_sum)
plt.show()
```

```python
##FROM ABOVE PLOT WE CAN CONCLUDE THAT MARICRUZ CASHIER IS HAVING HIGHER COLLECTION ###
```

```python
##FROM ABOVE PLOT WE CAN CONCLUDE THAT MARICRUZ CASHIER IS HAVING HIGHER COLLECTION ###
```

```python
sns.scatterplot(x='Amount',y='CashierName',data=df1_mean,hue="Received Cash")
```

```python
###  Median ###
df1_median=df.groupby(['CashierName'])['Amount','Received Cash'].median().reset_index()
```

```python
df1_median
```

```python
sns.relplot(x='Amount',y='CashierName',data=df1_median)
```

```python
###  Varience ###
df1_var=df.groupby(['CashierName'])['Amount','Received Cash'].var().reset_index()
```

```python
df1_var
```

```python
sns.relplot(x='Amount',y='CashierName',data=df1_var)
```

```python
###  Standard Deviation ###
df1_std=df.groupby(['CashierName'])['Amount','Received Cash'].std().reset_index()
```

```python
df1_std
```

```python
sns.relplot(x='Amount',y='CashierName',data=df1_std)
```
